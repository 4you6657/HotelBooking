```python
import jieba
import re
import nltk
from nltk import *
```

```python

# 正则对字符串清洗
def textParse(str_doc):
    # 正则过滤掉特殊符号、标点、英文、数字等。
    r1 = '[a-zA-Z0-9’!"#$%&\'()*+,-./:：;；|<=>?@，—。?★、…【】《》？“”‘’！[\\]^_`{|}~]+'
    # 去除空格
    r2 = '\s+'
    # 去除换行符
    str_doc=re.sub(r1, ' ', str_doc)
# 多个空格成1个
    str_doc=re.sub(r2, ' ', str_doc)
# 去除换行符
    # str_doc = str_doc.replace('\n',' ')
    return str_doc

# 读取文本信息
def readFile(path):
    str_doc = ""
    with open(path,'r') as f:
        str_doc = f.read()
    return str_doc

if __name__=='__main__':
    #读取文本
    path= r'D:\data\txt\nahan.txt'
    str_doc = readFile(path)
    #由于编码的问题，文本文件需要调整编码为ANSI才能够正常读出。
    #正则清洗字符串
    word_list=textParse(str_doc)
    print(word_list)
    import jieba
    #1\全模式,扫描所有可以成词的词语,速度非常快,不能解决歧义
    #jieba分词处理
    seg_list = jieba.cut(str_doc)
    print("===========================================")
    #词频统计
    fdist = FreqDist(seg_list)
    print("===========================================")
    print(fdist.keys(),fdist.values())
    w='王九妈'
    print(w,'出现频率:',fdist.freq(w)) #给定样本的频率
    print(w,'出现次数:',fdist[w])      #出现次数
    #频率分布表
    fdist=FreqDist(word_list)
    fdist.tabulate(10) #频率分布表
    
    #频率分布图
    
    
```
